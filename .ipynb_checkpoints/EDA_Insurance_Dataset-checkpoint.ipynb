# Feature Engineering 
- bmi column can be categorize into  underweight, normal ,overweight and obese 
- bloodpressure can be categorize into i-e normal, elevated, hbp-1 and hbp-2 and hyoertensive_crises

```python
def bmi_cat(value):
    if value<18.5: 
        return 'underweight'
    elif value>=18.5 and value<=25:
        return 'normal'
    elif value>=25 and value<=30:
        return 'overweight'
    else: 
        return 'obese'
```

```python
data['bmi_cat'] = data['bmi'].apply(lambda x: bmi_cat(x))
```

```python
# most individuals have normal weight
```

```python
plt.pie(data['bmi_cat'].value_counts(), autopct='%0.1f%%', labels=data['bmi_cat'].unique())
plt.show()
```

```python
def bloodpressure_cat(value):
    if value < 120:
        return "normal"
    elif value >= 120 and value < 130:
        return "elevated"
    elif value >= 130 and value < 140:
        return "hbp-stage1" # high blood pressure stage 1
    elif value >= 140 and value < 180:
        return "hbp-stage2" # high blood pressure stage 2
    elif value >= 180:
        return "hypertensive_crisis"
    else:
        return value
```

```python
data['bloodpressure_cat'] = data['bloodpressure'].apply(lambda x: bloodpressure_cat(x))
```

```python
# most individuals have normal bloodpressure
```

```python
plt.pie(data['bloodpressure_cat'].value_counts(), autopct='%0.1f%%', labels=data['bloodpressure_cat'].unique())
plt.show()
```

### Gender vs Claim
-  males have higher claims then females.Lets investigate more demographic features to explore the reason why is it that's so?

```python
temp_df = data.groupby('gender')['claim'].mean().reset_index()
```

```python
data.groupby('gender')['claim'].describe()
```

```python
sns.barplot(x='gender', y='claim', data=temp_df)
```

```python
df = data.groupby('gender')['bmi_cat'].value_counts().reset_index(name='Count')
df1= data.groupby('gender')['bloodpressure_cat'].value_counts().reset_index(name='Count')
df2 = data.groupby('gender')['smoker'].value_counts().reset_index(name='Count')
```

- Obesity and elevated blood pressure are risk factors that can contribute to higher healthcare costs.
- by looking at the graphs males are more likely to experience these conditions, that's why its reasonable to see higher claim amounts.
- Smoking is another significant risk factor for various health issues, we can see they are more male smokers

```python
# Create a figure with two subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 5))

# Plot the first bar plot on ax1
sns.barplot(x='bmi_cat', y='Count', hue='gender', data=df, ax=ax1)
ax1.set_xlabel('BMI Category')
ax1.set_ylabel('Count')

# Plot the second bar plot on ax2
sns.barplot(x='bloodpressure_cat', y='Count', hue='gender', data=df1, ax=ax2)
ax2.set_xlabel('Blood Pressure Category')
ax2.set_ylabel('Count')

sns.barplot(x='smoker', y='Count', hue='gender', data=df2, ax=ax3)
#ax2.set_xlabel('')
ax3.set_ylabel('Count')

# Adjust the layout to make sure the plots don't overlap
plt.tight_layout()

# Show the plot
plt.show()
```
